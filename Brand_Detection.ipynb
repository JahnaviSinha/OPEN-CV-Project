```python
%cd /content
import os


def update_labels(folder_path):
    # Get all label files in the directory
    labels_dir = os.path.join(folder_path, 'labels')
    for label_file in os.listdir(labels_dir):
        if label_file.endswith('.txt'):
            label_path = os.path.join(labels_dir, label_file)

            # Read the content of the label file
            with open(label_path, 'r') as f:
                lines = f.readlines()

            # Modify the first class (the first number in each line)
            new_lines = []
            for line in lines:
                parts = line.split()
                parts[0] = '0'  # Set the first class label to 0
                new_lines.append(' '.join(parts) + '\n')

            # Write the updated lines back to the file
            with open(label_path, 'w') as f:
                f.writelines(new_lines)

def process_grocery_dataset(base_path):
    # Process each folder (train, val, test)
    for folder in ['train', 'valid', 'test']:
        folder_path = os.path.join(base_path, folder)
        update_labels(folder_path)

# Base path of your dataset
base_path = '/content/Grocery_Product_Detection'
process_grocery_dataset(base_path)
```

```python
%cd /content/yolov5
import os

# Function to create custom.yaml
def create_yaml(dataset_dir, class_name, output_file='custom.yaml'):
    # Define the paths for train and validation (test)
    train_images_dir = os.path.join(dataset_dir, 'train/images')
    test_images_dir = os.path.join(dataset_dir, 'test/images')

    # Number of classes
    nc = 1  # Since you have only 1 class

    # YAML content
    yaml_content = f"""
# custom.yaml for YOLOv5
train: {train_images_dir}  # path to training images directory
val: {test_images_dir}     # path to validation (test) images directory

# Number of classes
nc: {nc}

# Class names
names: ['{class_name}']
"""

    # Write the content to a YAML file
    with open(output_file, 'w') as file:
        file.write(yaml_content)

    print(f"{output_file} has been created with the provided configuration.")

# Example usage
dataset_dir = '/content/Grocery_Product_Detection'  # Replace with actual path to your dataset
class_name = 'item'  # Replace with your actual class name,
create_yaml(dataset_dir, class_name)
```

```python
import numpy as np

def calculate_iou(bbox1, bbox2):
    """Calculate the Intersection over Union (IoU) of two bounding boxes."""
    x1 = max(bbox1[0], bbox2[0])
    y1 = max(bbox1[1], bbox2[1])
    x2 = min(bbox1[0] + bbox1[2], bbox2[0] + bbox2[2])
    y2 = min(bbox1[1] + bbox1[3], bbox2[1] + bbox2[3])

    # Compute the area of intersection
    intersection_area = max(0, x2 - x1) * max(0, y2 - y1)

    # Compute the area of both bounding boxes
    bbox1_area = bbox1[2] * bbox1[3]
    bbox2_area = bbox2[2] * bbox2[3]

    # Compute the union area
    union_area = bbox1_area + bbox2_area - intersection_area

    # Compute the IoU
    iou = intersection_area / union_area if union_area != 0 else 0
    return iou

def refine_annotations(annotations, iou_threshold=0.75):
    """Refine annotations by removing boxes with IoU > threshold and lower confidence."""
    refined_annotations = []
    annotations = sorted(annotations, key=lambda x: x['score'], reverse=True)  # Sort by confidence

    while annotations:
        current = annotations.pop(0)
        refined_annotations.append(current)
        annotations = [ann for ann in annotations if calculate_iou(current['bbox'], ann['bbox']) <= iou_threshold]

    return refined_annotations


refined_annotations = refine_annotations(annotation, iou_threshold=0.75)
```

```python
from matplotlib.patches import Rectangle
from PIL import Image
import matplotlib.pyplot as plt

def plot_image_with_bboxes(image_path, annotations):
    """Plot the image with bounding boxes."""
    image = Image.open(image_path)
    plt.figure(figsize=(10, 10))
    ax = plt.gca()

    for ann in annotations:
        bbox = ann['bbox']
        rect = Rectangle((bbox[0], bbox[1]), bbox[2], bbox[3],
                         linewidth=2, edgecolor='red', facecolor='none')
        ax.add_patch(rect)
        plt.text(bbox[0], bbox[1] - 5, f"{ann['category_name']} ({ann['score']:.2f})",
                 color='red', fontsize=12, backgroundcolor='white')

    plt.imshow(image)
    plt.axis('off')
    plt.savefig("/content/result.png")
    plt.show()


```

```python
import re
import os

# Example filename
filenames = os.listdir("/content/Brand_Recognition/test/images")
# print(filenames)

print(len(filenames))

def extract_brand_name(filename):
    match = re.match(r"^(.*?)-\d+-_jpg", filename)
    # print(match)
    if match:
        return match.group(1)
    else:
        return None

# Store the class mappings
class_mapping = {}
brand_images = {}


# Iterate over filenames and extract class names
for filename in filenames:
    # Remove the extension first
    file_base = os.path.splitext(filename)[0]
    # print(file_base)
    # Extract the brand name
    brand_name = extract_brand_name(file_base)
    # print(brand_name)
    if brand_name:
        # print(f"File: {filename} -> Brand: {brand_name}")
        # Add the brand name to class mapping
        if brand_name not in class_mapping:
            class_mapping[brand_name] = len(class_mapping) + 1
            brand_images[brand_name] = [filename]
        else:
          brand_images[brand_name].append(filename)

# Output the class mapping
print("\nClass Mapping:")
for brand, class_id in brand_images.items():
    print(f"Brand: {brand}, Length: {len(class_id)}")

```

```python
import os
import random
import shutil

def split_dataset_into_train_test(dataset_dir, train_dir, test_dir, train_ratio=0.8, seed=42):
    # Set random seed for reproducibility
    random.seed(seed)

    # Get all the brand directories (classes)
    brands = [d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))]

    # Shuffle the brands to split randomly
    random.shuffle(brands)

    # Split into train and test classes based on the train_ratio
    split_idx = int(len(brands) * train_ratio)
    train_brands = brands[:split_idx]
    test_brands = brands[split_idx:]

    # Function to copy files to a new directory
    def copy_files(brand_list, destination_dir):
        for brand in brand_list:
            brand_dir = os.path.join(dataset_dir, brand)
            images = os.listdir(brand_dir)

            # Create the brand directory in the destination (train/test)
            dest_brand_dir = os.path.join(destination_dir, brand)
            os.makedirs(dest_brand_dir, exist_ok=True)

            # Copy each image to the destination directory
            for img in images:
                shutil.copy(os.path.join(brand_dir, img), dest_brand_dir)

    # Copy files for training and testing
    copy_files(train_brands, train_dir)
    copy_files(test_brands, test_dir)

    print(f"Training brands: {train_brands}")
    print(f"Testing brands: {test_brands}")

# Example usage
dataset_dir = '/content/brand-recognition'
train_dir = '/content/dataset/train'
test_dir = '/content/dataset/test'

split_dataset_into_train_test(dataset_dir, train_dir, test_dir)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import datasets, transforms, models
import random
import cv2
from PIL import Image
import os
import torch.nn.functional as F

# DataLoader for Few-Shot Learning (Few-Shot Dataset Sampler)
class FewShotDataset(Dataset):
    def __init__(self, root_dir, num_classes_per_episode, num_support_per_class, num_query_per_class):
        self.root_dir = root_dir
        self.num_classes_per_episode = num_classes_per_episode
        self.num_support_per_class = num_support_per_class
        self.num_query_per_class = num_query_per_class

        # Create a mapping from class (folder) names to indices
        self.class_to_idx = {cls: idx for idx, cls in enumerate(os.listdir(root_dir))}
        self.samples = self.load_samples(root_dir)

    def load_samples(self, root_dir):
        samples = []
        # Iterate through each class directory
        for cls in os.listdir(root_dir):
            cls_dir = os.path.join(root_dir, cls)
            if os.path.isdir(cls_dir):  # Ensure it's a directory
                # Collect all images in this class directory
                for img_name in os.listdir(cls_dir):
                    img_path = os.path.join(cls_dir, img_name)
                    # Only add files with a specific image extension (e.g., .jpg, .png)
                    if img_path.endswith(('.jpg', '.jpeg', '.png')):
                        samples.append((img_path, self.class_to_idx[cls]))
        return samples

    def __getitem__(self, index):
        # Randomly sample `num_classes_per_episode` classes
        classes = random.sample(list(self.class_to_idx.keys()), self.num_classes_per_episode)
        support_images = []
        query_images = []
        support_labels = []
        query_labels = []

        # Create a label mapping for the selected classes
        label_mapping = {cls: i for i, cls in enumerate(classes)}

        for cls in classes:
            cls_idx = self.class_to_idx[cls]
            cls_images = [img for img, lbl in self.samples if lbl == cls_idx]

            # Ensure there are enough images for support and query
            if len(cls_images) < self.num_support_per_class + self.num_query_per_class:
                raise ValueError(f"Not enough images in class {cls} for support and query sampling.")

            # Randomly sample support and query images for this class
            selected_images = random.sample(cls_images, self.num_support_per_class + self.num_query_per_class)

            support_imgs = selected_images[:self.num_support_per_class]
            query_imgs = selected_images[self.num_support_per_class:]

            support_images.extend(support_imgs)
            query_images.extend(query_imgs)
            support_labels.extend([label_mapping[cls]] * self.num_support_per_class)
            query_labels.extend([label_mapping[cls]] * self.num_query_per_class)

        return support_images, query_images, support_labels, query_labels

    def __len__(self):
        return len(self.samples)  # Returns the number of samples in the dataset

# ResNet Backbone
class CustomCNNModel(nn.Module):
    def __init__(self):
        super(CustomCNNModel, self).__init__()

        # Define the layers of your CNN
        self.conv1 = nn.Conv2d(in_channels=3, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1)

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)

        # Define a fully connected layer
        self.fc1 = nn.Linear(32768, 256)  # Adjust based on the input image size
        self.fc2 = nn.Linear(256, 10)  # Example output size for classification (10 classes)

    def forward(self, x):
        # Apply convolutions and pooling
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))

        # Flatten the tensor
        x = x.view(x.size(0), -1)
        print(x.shape)
        # Fully connected layers
        x = F.relu(self.fc1(x))
        x = self.fc2(x)

        return x

class ResNetBackbone(nn.Module):
    def __init__(self):
        super(ResNetBackbone, self).__init__()
        resnet = models.resnet18(pretrained=True)
        self.backbone = nn.Sequential(*list(resnet.children())[:-1])  # Remove the final FC layer

    def forward(self, x):
        x = self.backbone(x)
        return x.view(x.size(0), -1)


# Prototypical Network
class PrototypicalNetwork(nn.Module):
    def __init__(self, backbone, embedding_dim):
        super(PrototypicalNetwork, self).__init__()
        self.backbone = backbone
        self.fc1 = nn.Linear(512, 512)
        self.fc3 = nn.Linear(512, embedding_dim)
        self.dropout = nn.Dropout(p=0.5)

    def forward(self, x):
        embeddings = self.backbone(x)
        embeddings = self.fc1(embeddings)
        embeddings = self.dropout(embeddings)
        embeddings = torch.relu(embeddings)
        embeddings = self.fc3(embeddings)
        return embeddings

    def compute_prototypes(self, support_embeddings, support_labels):
        support_labels = torch.tensor(support_labels)
        classes = torch.unique(support_labels)
        prototypes = torch.stack([support_embeddings[support_labels == c].mean(0) for c in classes])
        return prototypes

    def classify(self, query_embeddings, prototypes):
        distances = torch.cdist(query_embeddings.unsqueeze(0), prototypes.unsqueeze(0)).squeeze(0)
        return distances.argmin(dim=1)  # Class with the smallest distance




class PrototypicalLoss(nn.Module):
    def __init__(self):
        super(PrototypicalLoss, self).__init__()

    def forward(self, embeddings, targets, n_classes, n_support, n_query, query_labels, prototypes):
        query_labels = torch.tensor(query_labels, dtype=torch.long, device=embeddings.device)
        dists = torch.cdist(embeddings, prototypes)
        classifier = nn.CrossEntropyLoss()
        loss = classifier(-dists, query_labels)
        return loss

def train_prototypical_network(train_dir, test_dir):
    embedding_dim = 64
    num_classes_per_episode = 8
    num_support_per_class = 10
    num_query_per_class = 6
    num_epochs = 6
    train_batch_size = 1
    test_batch_size = 1
    test_classes_per_episode = 15
    test_support_per_class = 8
    test_query_per_class = 10

    transform = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor()
    ])

    train_dataset = datasets.ImageFolder(train_dir, transform=transform)
    test_dataset = datasets.ImageFolder(test_dir, transform=transform)

    train_loader = DataLoader(FewShotDataset(train_dir, num_classes_per_episode, num_support_per_class, num_query_per_class), batch_size=train_batch_size, shuffle=True)
    test_loader = DataLoader(FewShotDataset(test_dir, test_classes_per_episode, test_support_per_class, test_query_per_class), batch_size=test_batch_size, shuffle=True)

    backbone = ResNetBackbone().cuda() if torch.cuda.is_available() else ResNetBackbone()
    model = PrototypicalNetwork(backbone, embedding_dim).cuda() if torch.cuda.is_available() else PrototypicalNetwork(backbone, embedding_dim)
    optimizer = optim.Adam(model.parameters(), lr=1e-3, weight_decay=1e-5)
    criterion = PrototypicalLoss()

    for epoch in range(num_epochs):
        model.train()
        epoch_loss = 0
        num_episode = 250
        episode = 0

        for support_images, query_images, support_labels, query_labels in train_loader:
            if episode >= num_episode:
                break

            print(f"\rEpisode: {episode}/{num_episode}", end = " ")

            episode += 1
            optimizer.zero_grad()

            support_images = [torch.stack([transform(Image.open(img)).cuda() if torch.cuda.is_available() else transform(Image.open(img)) for img in batch]) for batch in support_images]
            query_images = [torch.stack([transform(Image.open(img)).cuda() if torch.cuda.is_available() else transform(Image.open(img)) for img in batch]) for batch in query_images]

            combined_images = torch.cat(support_images + query_images, dim=0)
            embeddings = model(combined_images)

            support_embeddings = model(torch.cat(support_images))
            query_embeddings = model(torch.cat(query_images))

            prototypes = model.compute_prototypes(support_embeddings, support_labels)

            query_labels = torch.tensor(query_labels, dtype=torch.long, device=embeddings.device)
            dists = torch.cdist(query_embeddings, prototypes)
            classifier = nn.CrossEntropyLoss()
            loss = classifier(-dists, query_labels)

            loss.backward()
            optimizer.step()

            epoch_loss += loss.item()

        print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss / num_episode}")

        model.eval()
        episode = 0
        num_samples = 0
        num_correct = 0
        num_episodes = 100

        #Testing After Each Epoch...
        for support_images, query_images, support_labels, query_labels in test_loader:
            if episode >= num_episode:
                break

            episode += 1
            support_images = [torch.stack([transform(Image.open(img)).cuda() if torch.cuda.is_available() else transform(Image.open(img)) for img in batch]) for batch in support_images]
            query_images = [torch.stack([transform(Image.open(img)).cuda() if torch.cuda.is_available() else transform(Image.open(img)) for img in batch]) for batch in query_images]

            support_embeddings = model(torch.cat(support_images))
            query_embeddings = model(torch.cat(query_images))

            prototypes = model.compute_prototypes(support_embeddings, torch.tensor(support_labels).cuda() if torch.cuda.is_available() else torch.tensor(support_labels))

            query_outputs = model.classify(query_embeddings, prototypes)
            num_samples += len(query_labels)
            num_correct += (query_outputs == torch.tensor(query_labels).cuda() if torch.cuda.is_available() else torch.tensor(query_labels)).sum().item()

        print(f"Testing Accuracy: {num_correct / (num_samples + 1)}")

    torch.save(model.state_dict(), 'prototypical_network_model.pth')


# Specify paths to the dataset directories
train_dir = '/content/dataset/train'
test_dir = '/content/dataset/test'

# Train the model
train_prototypical_network(train_dir, test_dir)
```
